```python
#Now split the dataframe in X and y
X=iris.iloc[:,:4]
X.head()
```

```python
y=iris.iloc[:,-1] #-1 mean last column of dataframe seleted and : mean all row selected
y.head()
```

Data Normalization

```python
#because sepal_lenght and sepat width different unit
X = preprocessing.StandardScaler().fit_transform(X)
X[0:4] 
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
#test_size=0.3 mean 70% data for training and 30% for testing
#random_state=1 mean when we run again and again same selection will used
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
y_test.shape #45 rows are randomly selected for testing
```

Training and Predicting

```python
knnModel=KNeighborsClassifier(n_neighbors=3) # 3 is value of K
knnModel.fit(X_train,y_train)
y_predict1=knnModel.predict(X_test)
```

Accuracy 

```python
# from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_predict1)
acc
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test.values,y_predict1)
cm
```

```python
cm1=pd.DataFrame(data=cm,index=['setosa','versicolor','virginica'],columns=['setosa','versicolor','virginica'])
cm1
```

```python
# Step 4: Select a machine learning algorithm
knn = KNeighborsClassifier(n_neighbors=3)  # Setting warn_on_dtype to False




```

```python
# Step 5: Train your model
knn.fit(X_train, y_train)

```

```python
# Step 6: Evaluate your model's performance
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
